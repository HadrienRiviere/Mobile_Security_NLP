```python
# Use this function to know what is inside the different folders
def update_folder_contents():
    manuals_pdf_folder = os.listdir(os.getcwd()+"/manuals_pdf")
    pdf_files = [f for f in manuals_pdf_folder if f.endswith('.pdf')]
    manuals_txt_folder = os.listdir(os.getcwd()+"/manuals_txt")
    txt_files = [f for f in manuals_txt_folder if f.endswith('.txt')]
    manuals_dict_folder = os.listdir(os.getcwd()+"/manuals_dict")
    dict_files = [f for f in manuals_dict_folder if f.endswith('.txt')]
    return pdf_files,txt_files,dict_files
```

```python
pdf_files, txt_files, dict_files = update_folder_contents()
```

```python
# Create corpus
text_data = []
for dict_file in dict_files : 
    with open('manuals_dict/'+dict_file,'r',encoding='utf-8') as filehandle:
        local_dict_2 = eval(filehandle.readline())
        text_data.append(local_dict_2)
print(text_data)
```

```python
# Create dictionary and corpus from the dict files : and save the model
dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
pickle.dump(corpus, open('model/corpus.pkl', 'wb'))
dictionary.save('model/dictionary.gensim')
```

```python
NUM_TOPICS = 2
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
ldamodel.save('model/model5.gensim')
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```

```python
with open('manuals_dict/'+"sonos_beam_dict.txt",'r',encoding='utf-8') as filehandle:
        new_doc = eval(filehandle.readline())
new_doc_bow = dictionary.doc2bow(new_doc)
print(ldamodel.get_document_topics(new_doc_bow))

# The print will show the probability of the new document to be in topic 0 or 1
```

```python
lda_display = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```

```python
dict_files_2 = ["amazon_echo_1st_gen_dict.txt","Amazon_Echo_Dot_3rdGen_QSG_US_sw_dict.txt","apple_tv_4thgen_sw_dict.txt",
              "eufy_lumos_bulb_dict.txt","facebook_portal_sw_dict.txt","google_home_dict.txt","google_home_mini_dict.txt",
              "google_homehub_sw_dict.txt","philips_hue_hub_dict.txt"]
text_data_2 = []
for dict_file in dict_files_2 : 
    with open('manuals_dict/'+dict_file,'r',encoding='utf-8') as filehandle:
        local_dict_3 = eval(filehandle.readline())
        text_data_2.append(local_dict_3)
print(text_data_2)
```

```python
dictionary_2 = corpora.Dictionary(text_data_2)
corpus_2 = [dictionary_2.doc2bow(text) for text in text_data_2]
pickle.dump(corpus_2, open('model/corpus_2.pkl', 'wb'))
dictionary_2.save('model/dictionary_2.gensim')
NUM_TOPICS = 2
ldamodel_2 = gensim.models.ldamodel.LdaModel(corpus_2, num_topics = NUM_TOPICS, id2word=dictionary_2, passes=15)
ldamodel_2.save('model/model2_2.gensim')
topics_2 = ldamodel_2.print_topics(num_words=4)
for topic in topics_2:
    print(topic)
with open('manuals_dict/'+"sonos_beam_dict.txt",'r',encoding='utf-8') as filehandle:
        new_doc = eval(filehandle.readline())
new_doc_bow = dictionary_2.doc2bow(new_doc)
print(ldamodel_2.get_document_topics(new_doc_bow))
```

```python
lda_display_2 = pyLDAvis.gensim.prepare(ldamodel_2, corpus_2, dictionary_2, sort_topics=False)
pyLDAvis.display(lda_display_2)
```
